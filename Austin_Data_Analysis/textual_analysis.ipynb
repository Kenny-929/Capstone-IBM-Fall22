```python
import rasterio
band_red = 1
band_green=2
band_blue=3
band_nir=4  

raster = rasterio.open('m_3009743_sw_14_060_20201018.tif')
red = raster.read(band_red) 
green=raster.read(band_green)
blue=raster.read(band_blue)
nir=raster.read(band_nir)
```

## Red Band Average

```python
red.shape
```

```python
import numpy as np

avg = np.zeros((red.shape[0]-2, red.shape[1]-2))

#slide a 3x3 window over the red bands
for i in range(red.shape[0]-2):
  for j in range(red.shape[1]-2):
    number= np.average([red[i][j],red[i][j+1],red[i][j+2],
                        red[i+1][j],red[i+1][j+1],red[i+1][j+2],
                        red[i+2][j],red[i+2][j+1],red[i+2][j+2]])
    avg[i][j]=number

#avg
```

```python
avg.shape
```

```python
print(avg)
```

## Green Band Average

```python
avg_green = np.zeros((green.shape[0]-2, green.shape[1]-2))

#slide a 3x3 window over the green bands
for i in range(green.shape[0]-2):
  for j in range(green.shape[1]-2):
    number_green= np.average([green[i][j],green[i][j+1],green[i][j+2],
                        green[i+1][j],green[i+1][j+1],green[i+1][j+2],
                        green[i+2][j],green[i+2][j+1],green[i+2][j+2]])
    avg_green[i][j]=number_green
```

```python
print(avg_green)
```

## Blue Band Average

```python
avg_blue = np.zeros((blue.shape[0]-2, blue.shape[1]-2))

#slide a 3x3 window over the blue bands
for i in range(blue.shape[0]-2):
  for j in range(blue.shape[1]-2):
    number_blue= np.average([blue[i][j],blue[i][j+1],blue[i][j+2],
                        blue[i+1][j],blue[i+1][j+1],blue[i+1][j+2],
                        blue[i+2][j],blue[i+2][j+1],blue[i+2][j+2]])
    avg_blue[i][j]=number_blue
```

```python
print(avg_blue)
```

## NIR Band Average

```python
avg_nir = np.zeros((nir.shape[0]-2, nir.shape[1]-2))

#slide a 3x3 window over the near-infrared bands
for i in range(nir.shape[0]-2):
  for j in range(nir.shape[1]-2):
    number_nir= np.average([nir[i][j],nir[i][j+1],nir[i][j+2],
                        nir[i+1][j],nir[i+1][j+1],nir[i+1][j+2],
                        nir[i+2][j],nir[i+2][j+1],nir[i+2][j+2]])
    avg_nir[i][j]=number_nir
```

```python
print(avg_nir)
```

## NDVI Average

```python
np.seterr(divide='ignore', invalid='ignore')
ndvi = (nir-red) / (nir+red)

avg_ndvi = np.zeros((nir.shape[0]-2, nir.shape[1]-2))

#slide a 3x3 window over ndvi value for all pixels
for i in range(nir.shape[0]-2):
  for j in range(nir.shape[1]-2):
    number_ndvi= np.average([ndvi[i][j],ndvi[i][j+1],ndvi[i][j+2],
                        ndvi[i+1][j],ndvi[i+1][j+1],ndvi[i+1][j+2],
                        ndvi[i+2][j],ndvi[i+2][j+1],ndvi[i+2][j+2]])
    avg_ndvi[i][j]=number_ndvi
```

```python
print(avg_ndvi)
```

## Convert to DataFrame

```python
import pandas as pd

red_df = pd.DataFrame(avg)
red_df.to_csv("red_band_average.csv")

green_df = pd.DataFrame(avg_green)
green_df.to_csv("green_band_average.csv")

blue_df = pd.DataFrame(avg_blue)
blue_df.to_csv("blue_band_average.csv")

nir_df = pd.DataFrame(avg_nir)
nir_df.to_csv("nir_band_average.csv")
```

```python
ndvi_df = pd.DataFrame(avg_ndvi)
ndvi_df.to_csv("ndvi_average.csv")
```
